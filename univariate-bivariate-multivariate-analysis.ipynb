# UNIVARIATE, BIVARIATE, MULTIVARIATE ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
dataframe.head(10)
```

```python
dataframe.shape
```

```python
## types of output classes/categories present in dataset
dataframe['species'].unique()
```

## 1. Univariate Analysis

```python
## DataFrame.loc ---> Access a group of rows and columns by label(s) or a boolean array.
dataframe_setosa = dataframe.loc[dataframe['species'] == 'setosa']
print(dataframe_setosa.head())
print(dataframe_setosa.shape)
```

```python
dataframe_versicolor = dataframe.loc[dataframe['species']=='versicolor']
print(dataframe_versicolor.head())
print(dataframe_versicolor.shape)
```

```python
dataframe_virginica = dataframe.loc[dataframe['species']=='virginica']
print(dataframe_virginica.head())
print(dataframe_virginica.shape)
```

## Observations ::
        50 species of setosa.
        50 species of versicolor.
        50 species of virginica.

```python
plt.plot(dataframe_setosa['sepal_length'],np.zeros_like(dataframe_setosa['sepal_length']),'o')
plt.title("-------------- UNIVARIATE ANALYSIS OF SEPAL_LENGTH -----------------")
plt.xlabel("sepal_length")
plt.ylabel("setosa")
plt.show()
```

```python
plt.plot(dataframe_versicolor['sepal_length'],np.zeros_like(dataframe_versicolor['sepal_length']),'o',color='red')
plt.title("-------------- UNIVARIATE ANALYSIS OF SEPAL_LENGTH -----------------")
plt.xlabel("sepal_length")
plt.ylabel("versicolor")
plt.show()
```

```python
plt.plot(dataframe_virginica['sepal_length'],np.zeros_like(dataframe_virginica['sepal_length']),'o',color='green')
plt.title("-------------- UNIVARIATE ANALYSIS OF SEPAL_LENGTH -----------------")
plt.xlabel("sepal_length")
plt.ylabel("versicolor")
plt.show()
```

## FOR SEPAL LENGTH

```python

plt.plot(dataframe_setosa['sepal_length'],np.zeros_like(dataframe_setosa['sepal_length']),'o',color='red')
plt.plot(dataframe_versicolor['sepal_length'],np.zeros_like(dataframe_versicolor['sepal_length']),'o',color='blue')
plt.plot(dataframe_virginica['sepal_length'],np.zeros_like(dataframe_virginica['sepal_length']),'o',color='green')
plt.xlabel('sepal_length')
plt.show()
```

## FOR SEPAL WIDTH

```python

plt.plot(dataframe_setosa['sepal_width'],np.zeros_like(dataframe_setosa['sepal_width']),'o',color='red')
plt.plot(dataframe_versicolor['sepal_width'],np.zeros_like(dataframe_versicolor['sepal_width']),'o',color='blue')
plt.plot(dataframe_virginica['sepal_width'],np.zeros_like(dataframe_virginica['sepal_width']),'o',color='green')
plt.xlabel('sepal_width')
plt.show()
```

## FOR PETAL LENGTH

```python

plt.plot(dataframe_setosa['petal_length'],np.zeros_like(dataframe_setosa['petal_length']),'o',color='red')
plt.plot(dataframe_versicolor['petal_length'],np.zeros_like(dataframe_versicolor['petal_length']),'o',color='blue')
plt.plot(dataframe_virginica['petal_length'],np.zeros_like(dataframe_virginica['petal_length']),'o',color='green')
plt.xlabel('petal_length')
plt.show()
```

## FOR PETAL WIDTH

```python

plt.plot(dataframe_setosa['petal_width'],np.zeros_like(dataframe_setosa['petal_width']),'o',color='red')
plt.plot(dataframe_versicolor['petal_width'],np.zeros_like(dataframe_versicolor['petal_width']),'o',color='blue')
plt.plot(dataframe_virginica['petal_width'],np.zeros_like(dataframe_virginica['petal_width']),'o',color='green')
plt.xlabel('petal_width')
plt.show()
```

## 2. BIVARIATE ANALYSIS

```python
## FacetGrid class helps in visualizing distribution of one variable 
## as well as the relationship between multiple variables 
## separately within subsets of your dataset using multiple panels.

sns.FacetGrid(dataframe,hue='species',size=6).map(plt.scatter,"petal_length","sepal_width").add_legend()
plt.show()
```
